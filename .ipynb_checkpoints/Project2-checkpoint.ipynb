### (문제) /data/boston.csv를 사용하여 아래와 같이 데이터를 탐색하고 요약하시오
**1)-1. 문제**  
데이터 셋에는 총 몇 개의 변수가 있고, 변수들의 이름은 어떻게 되며, 변수별로 결측값이 있는지 확인해보자. 

```python
# code for 1)-1
import pandas as pd
df = pd.read_csv('data/boston.csv')

# 총 15개의 변수와 이름을 확인할 수 있다.
print(df.columns.values.tolist())

# 변수별로 결측값 여부 확인
for i in df.columns:
    a = df[i].isnull().values.any()
    print(i,":", a)

# record 변수 index로 지정
df.set_index('record',inplace=True)
```

**1)-2. 문제**  
각 변수에 대해서 요약 통계량을 출력하시오.
어떤 변수의 변동이 가장 큰지, 어떤 변수가 치우쳐 있는지 요약 통계량을 바탕으로 설명하고, 치우쳐져 있는 변수 두 개를 선택하여 히스토그램을 나란히 그리고, 어떻게 치우쳐있는지 설명하시오 (변수는 정규화/표준화하지 않아도 된다.)


## record는 index 와 value가 같아 제외하면 'TAX','B' 변수의 std 값이 제일 크므로 이 둘의 값들이 각각의 평균으로 부터 많이 떨어져있어서 변수의 변동이 크다는 것을 알 수 있다. 표준편차가 크면 왜도의 절댓값이 커지므로 두 변수가 가장 한쪽으로 치우친다는 것을 알 수 있다. 

## 아래 히스토그램을 보면 'TAX' 변수는 200~400 사이에 변량들이 몰려있고 600 ~700사이에 변량들이 몰려있어 치우쳐진 곳이 2곳이다. 'B' 변수는 350 ~400 사이에 변량들이 치우쳐져 있다. 


```python
# code for 1)-2

# 변수별 요약 통계량 출력
print(df.describe().transpose())

# 'TAX','B' 의 히스토그램
import seaborn as sns
sns.displot(df['TAX'], bins=100)
sns.displot(df['B'])
```

**1)-3. 문제**  
모든 변수의 쌍의 상관계수를 구하고, heatmap을 이용하여 상관계수를 시각화하라.  
어떤 쌍이 가장 높은 상관관계를 나타내는지 heatmap을 근거로 찾으시오.

```python
# code for 1)-3
import matplotlib.pyplot as plt

cor = df.corr()
sns.heatmap(cor, annot=False,cmap=plt.cm.Blues)
plt.show()
```

**1)-5. 문제**   
상관계수값이 가장 큰 변수를 선택하여 해당 변수와 타깃 변수간의 관계를 시각화하기 위해 스캐터 플롯을 그려보라. 각 변수의 의미와 스캐터 플롯, 그리고 1)-4의 상관계수값을 연관지어 두 변수의 관계를 설명하라. 
(변수의 의미는 https://scikit-learn.org/stable/datasets/toy_dataset.html를 참고)

LSTAT 는 하위계층의 비율(%) 이고 MEDV는 본인 소유의 주택 가격의 중위수이다. 
이 두 변수의 상관계수 값은 -0.737663 으로 음수이다. 
아래 산점도의 분포를 보면 기울기가 전체적으로 감소하는 방향에 점들이 놓여있다.
따라서 LSTAT 와 MEDV 변수는 서로 반비례하며, 상관 관계 값이 음수이므로 음의 상관 관계에 있다. 또한 절댓값이 0.7이므로 1에 가까워 비교적 높은 상관 관계에 놓여있다고 볼 수 있다.
하위계층의 비율이 높을 수록 소유 주택 가격은 낮고, 하위계층의 비율이 낮을 수록 소유 주택가격이 높다는 것으로 보아,
상위 계층에 있을 수록 상대적으로 가격이 높은 집을 소유하고 있다는 것을 알 수 있다.

```python
# code for 1)-5

sns.lmplot(x='MEDV',y='LSTAT', data=df, fit_reg=False)
```
